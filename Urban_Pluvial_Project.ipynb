### Split Risk Lable & Then Drop Column

```python
import re #regex: to perform string operations

# --- Extract event_date (YYYY-MM-DD) ---
df['event_date'] = df['risk_labels'].str.extract(r'(\d{4}-\d{2}-\d{2})')

# --- Remove event_YYYY-MM-DD from labels ---
df['risk_labels_clean'] = df['risk_labels'].apply(
    lambda x: re.sub(r'event_\d{4}-\d{2}-\d{2}', '', x)
)

# --- Normalize labels ---
df['risk_labels_clean'] = (
    df['risk_labels_clean']
    .str.strip('|')                            # remove leading/trailing |
    .str.replace(r'\|+', '|', regex=True)      # collapse multiple pipes
    .str.lower()                               # make all lowercase
    .str.strip()                               # remove spaces around
)

# --- One-hot encode ---
risk_dummies = df['risk_labels_clean'].str.get_dummies(sep='|')

# --- Remove duplicate columns safely ---
risk_dummies = risk_dummies.T.groupby(level=0).max().T

# --- Merge back ---
df = pd.concat([df.drop(columns=['risk_labels_clean']), risk_dummies], axis=1)

df
```

```python
#Drop original risk_labels column
df=df.drop(columns=['risk_labels'])
```

```python
# Replace all NaN values in the 'event_date' column with the integer 0
df['event_date'] = df['event_date'].fillna(0)

```

### Split City_name & Then Drop Column

```python
# Step 1: Split the 'city_name' column by the comma and a space
df[['city', 'country']] = df['city_name'].str.split(', ', expand=True)

# Step 2: Remove all spaces from the new 'city' and 'country' columns
df['city'] = df['city'].str.replace(' ', '', regex=True)
df['country'] = df['country'].str.replace(' ', '', regex=True)
```

```python
df=df.drop(columns=['city_name'])
```

### New Column As is_risky

```python
# Define risk labels (excluding 'monitor')
risk_labels_filtered = ['ponding_hotspot', 'low_lying', 'sparse_drainage', 'extreme_rain_history']

# Add 'is_risky' column permanently
df['is_risky'] = (df[risk_labels_filtered].sum(axis=1) > 0).astype(int)
```

3. Clay Soil Regions (≈215 risky segments)

Resources:
1. High Rainfall Zones

    Since heavy rainfall is the most critical factor, the resources should focus on managing large volumes of water efficiently.

    * Improved Drainage Infrastructure: Invest in and maintain larger storm drains, culverts, and pumping stations to handle peak rainfall events.

    * Green Infrastructure: Promote the use of green spaces and natural systems to absorb water. This includes creating rain gardens, bioswales, and permeable pavements that allow water to soak into the ground instead of running off.

    * Early Warning Systems: Implement and expand weather forecasting and early warning systems to give residents and city officials time to prepare for high-intensity rain events.

2. Urbanized Areas

    The solutions for urbanized areas are about reducing the impact of high-density development on water runoff.

    * Building Codes and Zoning: Update urban planning policies to mandate the use of flood-resilient materials and designs in new construction. This includes requiring green roofs, rainwater harvesting systems, and ground-level permeable surfaces.

    * Retrofitting Existing Areas: Encourage residents and businesses to retrofit properties with features like rain barrels and permeable driveways.

    * Expansion of Green Spaces: Protect existing parks and open spaces and create new ones. These areas act as natural sponges, absorbing water and reducing the strain on the drainage system.

3. Clay Soil Regions

    While less critical, managing clay soil can significantly improve local-level flood resilience.

    * Soil Amendment Programs: Encourage the use of soil amendments such as compost or sand to improve the soil's drainage capacity.

    * Localized Stormwater Management: Implement small-scale solutions at the neighborhood or even household level. These could include creating swales (shallow channels) or installing French drains to redirect water away from structures.

#### Top recommendation for reducing flood risk in this regions (based on 7.4 analysis)
"HongKong", "Chennai", "Lagos", "Manila", "Barcelona", "Amsterdam", "Jakarta"

```python
# Focus only on selected cities
cities = ["HongKong", "Chennai", "Lagos", "Manila", "Barcelona", "Amsterdam", "Jakarta"]
sub = df[df['city'].isin(cities)]

# Summarize city-level metrics
summary = sub.groupby("city").agg(
    n_records=("segment_id","count"),
    risky_rate=("is_risky","mean"),
    avg_rainfall=("historical_rainfall_intensity_mm_hr","mean"),
    median_elev=("elevation_m","median"),
    avg_drainage_density=("drainage_density_km_per_km2","mean"),
    prop_low_lying=("low_lying","mean"),
    prop_ponding=("ponding_hotspot","mean"),
    prop_sparse_drainage=("sparse_drainage","mean"),
    prop_extreme_history=("extreme_rain_history","mean"),

    # Newly added
    most_common_soil=("soil_group", lambda x: x.mode()[0] if not x.mode().empty else None),
    avg_storm_drain_proximity=("storm_drain_proximity_m","mean"),
    most_common_drain_type=("storm_drain_type", lambda x: x.mode()[0] if not x.mode().empty else None),
    avg_return_period=("return_period_years","mean")
).reset_index()

# Thresholds for deciding recommendations
rain_mean = df['historical_rainfall_intensity_mm_hr'].mean()
drain_median = df['drainage_density_km_per_km2'].median()
elev_q25 = df['elevation_m'].quantile(0.25)
proximity_median = df['storm_drain_proximity_m'].median()
return_period_mean = df['return_period_years'].mean()

def top_recommend(row):
    # Drainage related
    if row['avg_drainage_density'] < drain_median or row['prop_sparse_drainage'] > 0.2:
        return "Upgrade & maintain drainage; add green infrastructure."
    
    # Rainfall & historical extremes
    if row['avg_rainfall'] > rain_mean or row['prop_extreme_history'] > 0.15:
        return "Increase stormwater storage/retention & improve early warning."
    
    # Low-lying areas
    if row['prop_low_lying'] > 0.2 or row['median_elev'] < elev_q25:
        return "Invest in flood defenses, pumps, and zoning in low-lying areas."
    
    # Ponding
    if row['prop_ponding'] > 0.15:
        return "Target ponding hotspots with local detention and drainage fixes."
    
    # Soil condition
    if row['most_common_soil'] in ["clay", "silty_clay", "sandy_clay"]:
        return "Soil has low permeability — use permeable pavements & rain gardens."
    
    # Storm drain proximity
    if row['avg_storm_drain_proximity'] > proximity_median:
        return "Improve storm drain network density & accessibility."
    
    # Storm drain type
    if row['most_common_drain_type'] in ["open", "surface"]:
        return "Upgrade open drains to covered/underground systems to improve flow."
    
    # Return period
    if row['avg_return_period'] >= return_period_mean:
        return "Plan for rare but extreme flood events — stronger infrastructure needed."
    
    # Default fallback
    return "Integrated approach: drainage, nature-based solutions, and emergency planning."

# Apply recommendations
summary["top_recommendation"] = summary.apply(top_recommend, axis=1)

# Display results neatly
pd.set_option("display.max_colwidth", None)
print(summary[[
    "city","n_records","risky_rate","avg_rainfall","median_elev","avg_drainage_density",
    "most_common_soil","avg_storm_drain_proximity","most_common_drain_type","avg_return_period",
    "top_recommendation"
]].to_string(index=False))
```

```python
# Create trigger flags (replace clay soil and open drain with actual values)
def trigger_flags(row):
    return pd.Series({
        "Low Drainage": int(row['avg_drainage_density'] < drain_median or row['prop_sparse_drainage'] > 0.2),
        "High Rainfall/Extreme": int(row['avg_rainfall'] > rain_mean or row['prop_extreme_history'] > 0.15),
        "Low Elevation/Low-Lying": int(row['prop_low_lying'] > 0.2 or row['median_elev'] < elev_q25),
        "Ponding Hotspots": int(row['prop_ponding'] > 0.15),
        "Soil Type": row['most_common_soil'],   # <-- keep actual soil type
        "Distant Drains": int(row['avg_storm_drain_proximity'] > proximity_median),
        "Drain Type": row['most_common_drain_type'],  # <-- keep actual drain type
        "Long Return Period": int(row['avg_return_period'] >= return_period_mean),
    })

# Apply to summary
triggers = summary.apply(trigger_flags, axis=1)
triggers["city"] = summary["city"]

# Split numeric (for heatmap) vs categorical (soil, drain type)
num_cols = ["Low Drainage","High Rainfall/Extreme","Low Elevation/Low-Lying",
            "Ponding Hotspots","Distant Drains","Long Return Period"]
cat_cols = ["Soil Type","Drain Type"]

# Heatmap for numeric triggers
plt.figure(figsize=(8,6))
sns.heatmap(triggers.set_index("city")[num_cols], annot=True, cmap="YlOrRd", cbar=False, linewidths=0.5)
plt.title("Flood Risk Trigger Factors (Numeric) by City")
plt.ylabel("City")
plt.xlabel("Trigger Factors")
plt.tight_layout()
plt.show()

# Show categorical factors separately as a clean table
print("\n=== Soil & Drain Types by City ===")
print(triggers[["city"] + cat_cols].to_string(index=False))
```
